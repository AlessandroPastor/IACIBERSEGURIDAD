## 2.1. Requisitos previos

Este proyecto usa un entorno virtual llamado `env` ubicado en la raíz del proyecto.

```bash
# Crear el entorno (ya ejecutado)
python -m venv env

# Activar en Windows
env\Scripts\activate

# Activar en Linux / macOS
source env/bin/activate
```

Asegúrate de estar ejecutando este notebook **dentro del kernel del entorno `env`**.

## 2.2. Verificación de la versión de Python

````python
import sys

version = sys.version_info
print(f"Python {sys.version}")
print("-" * 40)

if version.major == 3 and version.minor >= 9:
    print("[OK] Versión compatible (>= 3.9)")
else:
    print("[ADVERTENCIA] Se recomienda Python 3.9 o superior")
````

## 2.3. Instalación de bibliotecas

| Biblioteca | Función en seguridad |
|---|---|
| `numpy` | Operaciones matriciales sobre vectores de características |
| `pandas` | Carga y preprocesamiento de logs y datasets |
| `scikit-learn` | Algoritmos clásicos de ML (SVM, árboles, bosques) |
| `tensorflow` | Modelos de DL para detección de intrusiones |
| `keras` | API de alto nivel sobre TensorFlow |
| `pefile` | Análisis de ejecutables PE para detección de malware |
| `shap` | Explicabilidad de modelos (XAI) |
| `imbalanced-learn` | Manejo de clases desbalanceadas en datasets de seguridad |
| `matplotlib` | Visualización de resultados |
| `seaborn` | Visualización estadística avanzada |

````python
# Instalación de todas las dependencias del proyecto
import subprocess
import sys

paquetes = [
    "numpy",
    "pandas",
    "scikit-learn",
    "tensorflow",
    "matplotlib",
    "seaborn",
    "pefile",
    "requests",
    "joblib",
    "imbalanced-learn",
    "shap",
    "scipy",
]

print("Instalando dependencias...\n")
for pkg in paquetes:
    result = subprocess.run(
        [sys.executable, "-m", "pip", "install", pkg, "--quiet"],
        capture_output=True, text=True
    )
    estado = "[OK]" if result.returncode == 0 else "[ERROR]"
    print(f"{estado} {pkg}")

print("\nInstalación completada.")
````

## 2.4. Verificación del entorno

El siguiente script comprueba que todas las bibliotecas requeridas están disponibles e imprime su versión.

````python
import sys
import importlib

REQUIRED = [
    "numpy",
    "pandas",
    "sklearn",
    "tensorflow",
    "matplotlib",
    "seaborn",
    "pefile",
    "requests",
    "joblib",
    "imblearn",
    "shap",
    "scipy",
]

print(f"Python {sys.version}")
print("=" * 45)

todos_ok = True
for pkg in REQUIRED:
    try:
        mod = importlib.import_module(pkg)
        version = getattr(mod, "__version__", "OK")
        print(f"  [OK]     {pkg:<20} {version}")
    except ImportError:
        print(f"  [FALTA]  {pkg}")
        todos_ok = False

print("=" * 45)
if todos_ok:
    print("Entorno listo para los siguientes capítulos.")
else:
    print("Ejecuta la celda de instalación e intenta de nuevo.")
````
